```python
#Hacer encodings de las variables categoricas

from sklearn.preprocessing import OneHotEncoder

# Seleccionar las columnas categóricas
categorical_cols = ['Indice', 'Actividades Extracurriculares']

# Inicializar el codificador
encoder = OneHotEncoder(drop = 'first')

#Entrenar el codigicador con las columnas categóricas
encoder.fit('Completar el codigo')  

#Aplicar el codificador a las columnas categóricas
pd.DataFrame(encoder.transform(df[categorical_cols]).toarray(), columns=encoder.get_feature_names_out(categorical_cols))
```

```python
#Unimos el dataset al nuevo con las vartiables dummies
df = pd.concat([df, pd.DataFrame(encoder.transform(df[categorical_cols]).toarray(), columns=encoder.get_feature_names_out(categorical_cols),index=df.index)], axis=1)
```

```python
# Eliminamos las variables originales
df.drop(categorical_cols, axis=1, inplace=True)
```

```python
# 8. Identificación de outliers (por ejemplo, utilizando IQR)
Q1 = df['Practicas con Preguntas de Ejemplo'].quantile(0.25)
Q3 = df['Practicas con Preguntas de Ejemplo'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['Practicas con Preguntas de Ejemplo'] < Q1 - 1.5*IQR) | (df['Practicas con Preguntas de Ejemplo'] > Q3 + 1.5*IQR)]
print('Outliers:', outliers)
```
